## Exercício 01 - Escreva um programa Python para desenhar uma linha com rótulo adequado no eixo x, eixo y e um título.

```python
# Importar 
import matplotlib.pyplot as plt

x = range(1,50)
y = [value * 3 for value in x]

print('valores de x')
print (range(1,50))
print('valores de y(três vezes x )')
print(y)

plt.plot(x,y)

###
plt.title('Draw a line', fontsize = 15 , c="blue")

### Adicionar título ao eixo x
plt.xlabel('x - axis', fontsize = 12, c = "red")

### Adicionar título ao eixo y
plt.ylabel('y - axis', fontsize = 12, c= "black")


```
