```python
# Importar as bibliotecas

import matplotlib.pyplot as plt
import pandas as pd

# Mapear o caminho do documento

path= 'C:\\Users\\FIC\\CURSO PYTHON - FELIPE MUNHOS\\Arquivos\\'
file= 'fdata.csv'


#Abrir o documento
# O argumento index_col permite selecionar qual coluna será
# O indice do dataframe exemplo: 0 - primeira coluna como indice
# 1 - segunda coluna como indice e none - padrão

df= pd.read_csv(path + file, sep =',', index_col=None)

# Visualizar o cabelhaço

df.head()

# montar o grafico 

df.plot()
```

```python
df.head()

df.columns = ['data','abertura','alta','baixa','fechamento']

df.head()
```

```python
# Selecionando colunas do dataframe

df_2 = df[['data',"abertura"]]

df_2.plot()
```

```python
# Converter em numerico pf.tonumerci(coluna)

# Substituir um string
df_2['abertura'] = df_2['abertura'].replace('774. 25',"774.25")
df_2['abertura'] 

# Converter em numerico pd.to_numeric (coluna_)
df_2 ['abertura'] = pd.to_numeric(df_2['abertura'])

df_2.plot()
```
