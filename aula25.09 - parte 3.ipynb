```python
import matplotlib.pyplot as plt
import numpy as np

# Retorna (Linspace) valores_igualmente espaçodos em um intervalo Neste exemplo, vamos plotar duas curvas em um único gráfico.
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
plt.plot (x,y)
plt.plot (x,z)
```

```python
# O comando plt.subplot permite criar grupo de gráfico
# plt.subplot (linhas, Colunas, Posição elemento)

plt.subplot(2,1,1) # Primeiro gráfico
plt.plot (x,y) # Conjunto de dados do primeiro grafico
plt.title('subplots')
plt.subplot(2,1,2) # segundo gráfico
plt.plot(x,z ) # Conjunto e dados do segundo gráfico
```

```python
# O comando plt.subplot permite criar grupo de gráfico
# plt.subplot (linhas, Colunas, Posição elemento)

plt.subplot(1,2,1) # Primeiro gráfico
plt.plot (x,y) # Conjunto de dados do primeiro grafico
plt.title('subplots')
plt.subplot(1,2,2) # segundo gráfico
plt.plot(x,z ) # Conjunto e dados do segundo gráfico
```
