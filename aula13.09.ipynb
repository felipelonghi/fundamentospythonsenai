# l) Gere um arquivo em Excel para representar as vendas efetuadas pela filial norte em 2010.

```python
#df[df(['data']>='2005-10-08') & (df['data']<'2005-10-17') & (df['filial']=="norte")]

dados = df [(df['Data']>='2010-01-01') &
          (df['Data']<='2010-12-31') &
          (df['filial']== 'Norte')
           ]

# nome do arquivo

nome_arquivo = "norte_2010.xlsx"

# Salvar / Gerar um arquivo em excel

dados.to_excel(pasta_destino + nome_arquivo)
```

# m) Gere um arquivo em Excel para representar as vendas efetuadas pela filial Sudeste em 2010. Sendo que no relatório deve conter apenas vendas de casas sem garagem.

```python
dados = df[(df ['tipo'] =="Casa") & 
         (df["garag."]==0) &
         (df['filial'] =="sudeste") &
         (df['Data']>='2010-01-01')&
         (df['Data']<="2010-12-31")
         ]    

# nome do arquivo       
nome_arquivo = "sudeste_2010_Casa_SemGarag.xlsx"

# Salvar / Gerar um arquivo em excel

dados.to_excel(pasta_destino + nome_arquivo)
```
