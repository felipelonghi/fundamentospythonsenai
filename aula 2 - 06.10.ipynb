```python
x = protetorsolar[['temperatura']]
y = protetorsolar["qtde"]
```

### PASSO 05 - MODELO TREINO
IMPORTAR A BIBLIOTECA SCIKIT LEARN E DIVIDIR AS BASES EM TREINOS E TESTE, SENDO QUE O TAMANHO DO TESTE SERÁ DE 25% 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split (x,y,test_size=0.5)
```

```python
X_train.shape
```

### Importar o algoritmo LinearRegression da biblioteca Scikitlearn

```python
from sklearn.linear_model import LinearRegression
```

```python
# Este comando irá gerar a linha de tendência
# E calcular os coefientes linear e angular

regressor = LinearRegression(fit_intercept = True )

regressor.fit(X_train,y_train)
```

```python
print('Coeficiente angular (a):', regressor.coef_)
print('Coeficiente linear (b):', regressor.intercept_)
```

```python
y_previsao = regressor.predict(X_test)
y_previsao 
```

```python
plt.scatter(X_train,y_train,color = "blue")
plt.plot(X_train,regressor.predict(X_train),color = 'red')
plt.ylabel('Receita[Dólares]')
plt.xlabel('Temperatura [ºC]')
plt.title('Receita gerada vs. Temperatura (Protetor solar)')
```
