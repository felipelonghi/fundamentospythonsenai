```python
# Importar a biblioteca do Pand

import pandas as pd

caminho = 'C:\\Users\\FIC\\CURSO PYTHON - FELIPE MUNHOS\\016 - Python -  Fundamentos de Big Data e Data Analytics-20230911T220810Z-001 (2)\\016 - Python -  Fundamentos de Big Data e Data Analytics\\003 - Arquivos\\'
arquivo = "dados-vendas.xlsx"

# Atribuir os dados deste arquivo, em um dataframe

base = pd.read_excel(caminho + arquivo, sheet_name= "Vendas")

# Exibir os cabeçalhos das colunas

base.head()
```

```python
# Criando a tabela fato com as colunas:

df_dados = base [["ID-Produto","Loja","ID-Vendedor","Data Venda","ValorVenda"]]

# Renomera a coluna loja para ID loja

df_dados = df_dados.rename(columns= {"Loja": 'ID-Loja'})

df_dados.head()

# Salvar o arquivo na pasta dados- vandas

caminho_novo  = caminho + 'dados_vendas\\'
arquivo = "tabela-fato.xlsx"

df_dados.to_excel(caminho_novo + arquivo, index = False)
```

```python
# Criando a tabela fato com as colunas:

df_dados = base [["ID-Produto","Produto","Categoria","Segmento","Fabricante"]]

# Renomera a coluna loja para ID loja

df_dados = df_dados.rename(columns= {"Loja": 'ID-Produto'})

df_dados.head()

# Salvar o arquivo na pasta dados- vandas

caminho_novo  = caminho + 'dados_vendas\\'
arquivo = "tabela-fato.xlsx"

df_dados.to_excel(caminho_novo + arquivo, index = False)
```
