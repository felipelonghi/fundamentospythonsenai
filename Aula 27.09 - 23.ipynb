```python
grafico = sns.barplot(x='Segmento',y='ValorVenda',data=df)
# O metodo sns.barplot vai retornar uma Lista de sub
# metodos utilizando os containers para acessar o testo
# dos rotulos para cada coluna, desta forma o python entende
# que ele deve adicionar um rótulo em cada coluna, de acordo com o comando graficos.containers[0]

grafico.bar_label(grafico.containers[0])
```
