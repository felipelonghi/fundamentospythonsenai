```python
# O alias ( Apleido) da biblioteca Seaborn Fica com sns

import seaborn as sns
import numpy as np

# Retorna (Linspace) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100) 
y = np.sin(x) 
```
